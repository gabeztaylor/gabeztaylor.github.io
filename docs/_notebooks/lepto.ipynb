```python
import numpy as np
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
```

```python
n = 100000
data = np.zeros(n)
a, b = 0.4, 2.6
regions = np.zeros(n)
for i in range(n):
    region = np.random.choice([-1, 0, 1], p = [1/4, 1/2, 1/4])
    regions[i] = region
    if region < 0:
        data[i] = np.random.uniform(low = -b, high = -a)
    elif region == 0:
        data[i] = np.random.uniform(low = -a, high = a)
    else:
        data[i] = np.random.uniform(low = a, high = b)
```

```python
plt.hist(data)
plt.show()
```

```python
data.std()
```

```python
kurtosis(data)
```
